```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel(r'default of credit card clients.xls',header=None)
```

```python
df.head()
```

```python
df.iloc[1]
```

```python
headers = df.iloc[1]
new_df  = pd.DataFrame(df.values[1:], columns=headers)
```

```python
new_df
```

```python
newDf = new_df.drop(0, axis=0)
newDf.head()
```

```python
newDf.columns=newDf.columns.str.lower()
```

```python
newDf=newDf.rename(columns={'default payment next month':'default_payment'})
```

```python
newDf
```

```python
col=['id', 'limit_bal', 'sex', 'education', 'marriage', 'age', 'pay_0',
       'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6', 'bill_amt1', 'bill_amt2',
       'bill_amt3', 'bill_amt4', 'bill_amt5', 'bill_amt6', 'pay_amt1',
       'pay_amt2', 'pay_amt3', 'pay_amt4', 'pay_amt5', 'pay_amt6','default_payment']
```

```python
newDf=newDf[col].astype(int)
newDf.dtypes
```

```python
X_cols=['id', 'limit_bal', 'sex', 'education', 'marriage', 'age', 'pay_0',
       'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6', 'bill_amt1', 'bill_amt2',
       'bill_amt3', 'bill_amt4', 'bill_amt5', 'bill_amt6', 'pay_amt1',
       'pay_amt2', 'pay_amt3', 'pay_amt4', 'pay_amt5', 'pay_amt6']
y_cols=['default_payment']
```

```python
X=newDf[X_cols]
X
```

```python
y=newDf[y_cols]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)

X_train
y_train
```

```python
from sklearn import linear_model
log_model = linear_model.LogisticRegression()
log_model.fit(X = X_train, 
              y = y_train)
```

```python
y_pred=log_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
